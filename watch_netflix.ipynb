```python
%%writefile wantbacon.py
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
X1 = np.linspace(5, 60, 56)
C1 = 100*X1/X1
X2 = np.linspace(60, 95, 36)
C2 = 250 - X2*2.5
with plt.xkcd():
    plt.title('XKCD style plot!!!')
    plt.plot(X1, C1, 'cyan')
    plt.plot(X2, C2, 'cyan')
plt.xlabel('time'+'\n'+'Stove Ownership\" from xkcd by Randall Monroe')
plt.ylabel('my overall data plan')
plt.annotate(r'THE DAY I REALIZED'+'\n'+'I COULD WATCH NETFLIX'+'\n'+'WHENEVER I WAS FREE', xy = (60, 100), xycoords='data', xytext=(0.2, 0.8),
             textcoords = 'axes fraction', fontsize =8,
             arrowprops=dict(arrowstyle="->", linewidth = 1, color='k'))
plt.show()
```
